```python
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
souce=requests.get('https://en.wikipedia.org/wiki/List_of_countries_by_freshwater_withdrawal')
souce
```

```python
source=BeautifulSoup(souce.text)
source
```

```python
data1=[]
#data2=[]
data3=[]
data4=[]
data5=[]
data6=[]
data7=[]
data8=[]
```

```python
for i in source.find_all('tr'):
    col=i.find_all('td')
    if len(col)==8:
        data1.append(col[0].find(text=True))
        #data2.append(col[1].find(text=True))
        data3.append(col[2].find(text=True))
        data4.append(col[3].find(text=True))
        data5.append(col[4].find(text=True))
        data6.append(col[5].find(text=True))
        data7.append(col[6].find(text=True))
        data8.append(col[7].find(text=True))
```

```python
source1=source.find(class_='wikitable sortable')
item1=source1.find_all('a')
data9=[]
for i in item1:
    data=i.text
    data9.append(data)
len(data9)
```

```python
df=pd.DataFrame()
```

```python
df['Rank']=data1
df['Country']=data9
df['Totalwithdrawal(km³/year)']=data3
df['Usage_of_water']=data4
df['Domesticwithdrawal(%)']=data5
df['Industrialwithdrawal(%)']=data6
df['Agriculturalwithdrawal(%)']=data7
df['Date of  Information']=data8
```
